```python
import json
import pandas as pd

# Load the dataset
with open('train-v1.1.json', 'r') as f:
    squad_data = json.load(f)

# Explore the dataset
print(squad_data.keys())
print(squad_data['version'])
print(squad_data['data'][0])
```

```python
contexts = []
questions = []
answers = []

for entry in squad_data['data']:
    for paragraph in entry['paragraphs']:
        context = paragraph['context']
        for qa in paragraph['qas']:
            question = qa['question']
            for answer in qa['answers']:
                contexts.append(context)
                questions.append(question)
                answers.append(answer)

df = pd.DataFrame({'context': contexts, 'question': questions, 'answer': answers})
print(df.head())
```

**METRICS**

```python
def exact_match(predicted_answer, true_answer):
    return predicted_answer.strip().lower() == true_answer.strip().lower()

predicted_answer = "Paris"
true_answer = "paris"
print(exact_match(predicted_answer, true_answer))
```

```python
from sklearn.metrics import f1_score

def f1_score_qa(predicted_answer, true_answer):
    predicted_tokens = predicted_answer.strip().lower().split()
    true_tokens = true_answer.strip().lower().split()

    common_tokens = set(predicted_tokens) & set(true_tokens)

    if len(predicted_tokens) == 0 or len(true_tokens) == 0:
        return 0

    precision = len(common_tokens) / len(predicted_tokens)
    recall = len(common_tokens) / len(true_tokens)

    if precision + recall == 0:
        return 0
    return 2 * (precision * recall) / (precision + recall)

predicted_answer = "the capital of France is Paris"
true_answer = "Paris is the capital of France"
print(f1_score_qa(predicted_answer, true_answer))
```

```python
# Example dataset
questions = ["What is the capital of France?", "Who wrote Hamlet?"]
true_answers = ["Paris", "William Shakespeare"]
predicted_answers = ["Paris", "Shakespeare"]

# Initialize metrics
total_em = 0
total_f1 = 0

# Evaluate each question
for pred, true in zip(predicted_answers, true_answers):
    total_em += exact_match(pred, true)
    total_f1 += f1_score_qa(pred, true)

# Calculate averages
average_em = total_em / len(questions)
average_f1 = total_f1 / len(questions)

print(f"Average Exact Match (EM): {average_em * 100:.2f}%")
print(f"Average F1 Score: {average_f1 * 100:.2f}%")
```

```python
from transformers import pipeline

# Load the QA pipeline
qa_pipeline = pipeline("question-answering", model="bert-large-uncased-whole-word-masking-finetuned-squad")
```

```python
# Example context and question
context = """
The Eiffel Tower is a wrought-iron lattice tower on the Champ de Mars in Paris, France.
It is named after the engineer Gustave Eiffel, whose company designed and built the tower.
"""
question = "Who designed the Eiffel Tower?"

# Get the answer
result = qa_pipeline(question=question, context=context)
print(f"Question: {question}")
print(f"Answer: {result['answer']}")
print(f"Confidence: {result['score']:.2f}")
```

```python
def ask_question():
    # Input context
    context = input("Enter the context: ")

    # Input question
    question = input("Enter your question: ")

    # Get the answer
    result = qa_pipeline(question=question, context=context)
    print(f"\nQuestion: {question}")
    print(f"Answer: {result['answer']}")
    print(f"Confidence: {result['score']:.2f}\n")

# Run the QA system in a loop
while True:
    ask_question()
    another = input("Do you want to ask another question? (yes/no): ").lower()
    if another != "yes":
        break
```

```python
from transformers import pipeline

qa_pipeline = pipeline("question-answering", model="bert-large-uncased-whole-word-masking-finetuned-squad")
```

```python
import gradio as gr

def answer_question(question, context):
    result = qa_pipeline(question=question, context=context)
    return result['answer']

interface = gr.Interface(
    fn=answer_question,
    inputs=["text", "text"],
    outputs="text",
    title="Question Answering System",
    description="Ask a question and provide a context to get the answer.",
    examples=[
        ["Who designed the Eiffel Tower?", "The Eiffel Tower is a wrought-iron lattice tower on the Champ de Mars in Paris, France. It is named after the engineer Gustave Eiffel, whose company designed and built the tower."],
        ["What is the capital of France?", "France is a country in Europe. Its capital is Paris."]
    ]
)

interface.launch()
```
